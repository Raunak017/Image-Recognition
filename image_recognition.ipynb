```python
from torchvision import models
dir(models)
```

```python
alexnet = models.AlexNet()
```

```python
resnet = models.resnet101(pretrained=True)
resnet
```

```python
from torchvision import transforms
preprocess = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )])
```

```python
%matplotlib inline
import glob
from IPython.display import Image, display
for imageName in glob.glob('images/*.jpg'): #assuming JPG
    display(Image(filename=imageName))
    print(imageName)
```

```python
from PIL import Image as PIM
image_list = []
for filename in glob.glob('images/*.jpg'): #assuming gif
    im=PIM.open(filename)
    image_list.append(im)

image_list
```

```python
img_t1 = preprocess(image_list[0])
img_t2 = preprocess(image_list[1])
img_t3 = preprocess(image_list[2])
img_t4 = preprocess(image_list[3])
```

```python
import torch
batch_t1 = torch.unsqueeze(img_t1, 0)
resnet.eval()
out1 = resnet(batch_t1)

batch_t2 = torch.unsqueeze(img_t2, 0)
resnet.eval()
out2 = resnet(batch_t2)

batch_t3 = torch.unsqueeze(img_t3, 0)
resnet.eval()
out3 = resnet(batch_t3)

batch_t4 = torch.unsqueeze(img_t4, 0)
resnet.eval()
out4 = resnet(batch_t4)
out4
```

```python
with open('Imagenet.txt') as f:
    display(Image(filename='./images/image1.jpg'))
    labels = [line.strip() for line in f.readlines()]
    _, index = torch.max(out1, 1)
    
    percentage1 = torch.nn.functional.softmax(out1, dim=1)[0] * 100
    print(labels[index[0]], percentage1[index[0]].item())
```
